# Test of the Navarro & Fuss first passage time approximation

## General remarks
If we let X(t) denote the "state of evidence" at time t, this can be modelled with a diffusion process: a simple dynamic system in which the state of evidence evolves via the stochastic differential equation
d/dt X(t) = Normal(mu, sig^2).

The initial state of evidence X(0) lies in the range 0 < X(0) < a, the model predicts that a decision is made at the first time t for which X(t)<0 or X(t)>a, where absorption at the lower-boundary
(X(t)=0) corresponds to one possible choice and the absorption at the upper boundary (X(t)=a) results in the other choice.

Statistically, if we fix the Wiener process variance1 at sig^2 = 1, we refer to the resulting distribution as the Wiener first-passage time (WFPT) distribution.

```python
# useful libraries
from math import sqrt
from scipy.stats import norm
import numpy as np
import pylab as pl
```

Here is the implementation of their method

```python
def wfpt(t, v, a, w = None, err = None):
    # First passage time for Wiener diffusion model. 
    
    # Approximation based on Navarro & Fuss (2009).
    # USAGE: p = wfpt(t,v,a,w,err)
    # INPUTS:
    #   t - hitting time (e.g., response time in seconds)
    #   v - drift rate
    #   a - threshold
    #   w - bias (default: 0.5)
    #   err - error threshold (default: 1e-6)
    #
    # OUTPUTS:
    #   p - probability density
    
    if w is None:
        w = 0.5
    if err is None:
        err = 1e-6
    
    P = np.zeros(len(t))
    
    for i in range(len(t)):
        tt = t[i]/(a**2) # use normalized time

        # calculate number of terms needed for large t
        if np.pi*tt*err<1:                                      # if error threshold is set low enough
            kl = np.sqrt(-2*np.log(np.pi*tt*err)/(np.pi**2*tt)) # bound
            kl = np.max([kl,1/(np.pi*np.sqrt(tt))])             # ensure boundary conditions met
        else:                                                   # if error threshold set too high
            kl=1./(np.pi*np.sqrt(tt))                           # set to boundary condition

        # calculate number of terms needed for small t
        if 2*np.sqrt(2*np.pi*tt)*err<1:                         #  if error threshold is set low enough
            ks = 2 + np.sqrt(-2*tt*np.log(2*np.sqrt(2*np.pi*tt)*err))  # bound
            ks = np.max([ks,np.sqrt(tt)+1])                     # ensure boundary conditions are met
        else:                                                   # if error threshold was set too high
            ks = 2                                              # minimal kappa for that case

        # compute f(tt|0,1,w)
        p=0;                                                    # initialize density
        if ks<kl:                                               # if small t is better...
            K = np.ceil(ks)                                     # round to smallest integer meeting error
            for k in np.arange(-np.floor((K-1)/2), np.ceil((K-1)/2), 1): # loop over k
                p = p + (w+2*k)*np.exp(-((w+2*k)**2)/2/tt);     # increment sum

            p = p/np.sqrt(2*np.pi*tt**3);                       # add constant term
        else:                                                   # if large t is better...
            K = np.int(np.ceil(kl))                             # round to smallest integer meeting error
            for k in range(K):
                p = p + k*np.exp(-(k**2)*(np.pi**2)*tt/2)*np.sin(k*np.pi*w); # increment sum

            p=p*np.pi;                                          # add constant term

        # convert to f(t|v,a,w)
        P[i] = p*np.exp(-v*a*w -(v**2)*t[i]/2)/(a**2);

    return P
```

# Implementation of 1-d Brownian motion with drift

```python
# Brownian motion wrapped in a function
def brownian(x0, T, dt, delta, mu, samples):
    """
    see wikipedia.
    
    PARAMETERS:
        x0: initial conditions
        T: length of simulation
        dt: timestep
        delta: diffusion variable, variance(t) = delta**2 *t
        mu: drift
        samples: number of independent sample of this brownian process
    """
    times = np.arange(0, T, dt) 
    Nt = times.size
    stdev = sqrt(dt)*delta
    X = np.zeros((Nt,samples)) + x0
    for it in range(1, Nt):
        X[it,:] = X[it-1,:] + np.random.normal(mu*dt, stdev, samples)
    return X, times
```

```python
samples = 7
s, times = brownian(x0 = 0.5, T = 5, dt = 0.0001, delta = 1, mu = .7, samples = samples)
```

```python
def plot_paths(s, times):
    samples = s.shape[1]
    for ip in range(samples):
        pl.plot(times, s[:,ip])
    pl.xlabel('time, t')
    pl.ylabel('distance/decision variable, x ')

plot_paths(s, times)
```

# Calculate first passage time

```python
# Calculate first passage of a timeseries, wrapped in a function
# this is done for "standard" choice, X leaving the range [0,threshold] 

def first_passage(timeseries, threshold, times):
    event = (timeseries>threshold) + (timeseries<0)       # detect if a threshold crossing happened, 
    if sum(event)>0:                                      # i.e. >threshold or <0. "or" with python's "+"
        t = times[np.argmax(event)]                       # if this happened, find it.
    else:
        t = np.inf
    return t
```

```python
def plot_paths_crossings(s, times, Ts):
    samples = s.shape[1]
    for ip in range(samples):
        plotrange = times<Ts[ip]
        xs = s[plotrange,ip]
        pl.plot(times[plotrange], xs)
        if xs[-1]>0.5:
            pl.plot(Ts[ip],0.99,'ro')
        else:
            pl.plot(Ts[ip],0.01,'bo')
    pl.xlabel('time, t')
    pl.ylabel('distance/decision variable, x ')

plot_paths_crossings(s, times, Ts)
```

# Test of Sam's implementation

First, let's compare a random choice of parameters and the histogram.

```python
# Parameters
Nsamples = 1024
T = 5 # length of simulation
dt = 0.001  # only for resolution
threshold = 1
x0 = 0.5  # "bias"
mu = 0    # "drift"

# Get data from simulations
s, times = brownian(x0, T, dt, 1, mu, samples = Nsamples)

# Calculate the first passage time
Ts = np.array([first_passage(s[:,i], threshold, times = times) for i in range(Nsamples)])

# Get theory prediction
tbins = np.arange(0.01,max(Ts),0.01)
probs = wfpt(tbins, v = mu, a = threshold, w = x0);

# Plot both together
pl.hist(Ts, tbins, density = True)
pl.plot(tbins, probs, 'r--')
pl.xlabel('Waiting time, t')
pl.ylabel('Prob. density over simulations [blue] / analytical result [red], x ')
```

They seem pretty similary. To better compare, it might be useful to plot cumulative distribution functions, and vary one of the parameters, e.g. drift

Here we plot the cumulative distribution of waiting times for two different drift values

```python
from scipy.stats import cumfreq


Nsamples = 1024
threshold = 1
x0 = 0.5  # "bias"

for mu in [0, 3]:
    if mu == 0:
        dt = 0.0001
        T = 5
        m = 'r'
    if mu == 3:
        dt = 0.0001
        T = 1
        m = 'g'

    s, times = brownian(x0, T, dt, 1, mu, samples = Nsamples)
    Ts = np.array([first_passage(s[:,i], threshold, times = times) for i in range(Nsamples)])
    Ts = Ts[np.isfinite(Ts)]
    tbins = np.arange(0.01,max(Ts),0.01)
    probs = wfpt(tbins, v = mu, a = threshold, w = x0);
    a = cumfreq(Ts, numbins = len(tbins))
    pl.plot(tbins, a.cumcount/Nsamples, m, marker = 'o', linestyle = 'None')
    pl.plot(tbins, np.cumsum(probs)/np.sum(probs), 'k--', linewidth = 1.5)
    
pl.xlabel('Waiting time, t')
pl.ylabel('CUMULATIVE prob. density, model vs. simulation ')
```
